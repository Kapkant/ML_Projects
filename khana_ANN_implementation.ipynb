```python
#Import Packages
import numpy as np
import pandas as pd
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score,recall_score, f1_score
```

```python
#Using Kaggle, dataset for cancer
dataset = pd.read_csv('/kaggle/input/breast-cancer-wisconsin-data/data.csv')
```

```python
#inep and dep vars assigned
X = dataset.iloc[:,2:32].values
y = dataset.iloc[:,1].values
```

```python
#Label encoder
from sklearn.preprocessing import LabelEncoder, StandardScaler
labelencoder_Y=LabelEncoder()
y= labelencoder_Y.fit_transform(y)
```

```python
#test data set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 0)
```

```python
scale_X= StandardScaler()
X_train = scale_X.fit_transform(X_train)
X_test = scale_X.transform(X_test)
```

```python
#importing key ML Packages
import keras
from keras.models import Sequential
from keras.layers import Dense,Dropout

classifier = Sequential()
```

```python
classifier.add(Dense(units=16,kernel_initializer="uniform",
                    activation="relu",input_dim=30))
classifier.add(Dense(units=16,kernel_initializer="uniform",activation="relu"))
classifier.add(Dense(units=16,kernel_initializer="uniform",activation="relu"))
classifier.add(Dense(units=1,kernel_initializer="uniform",activation = "sigmoid"))
classifier.add(Dropout(rate=0.1))
```

```python
classifier.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
classifier.fit(X_train,y_train,batch_size=100,epochs=150)
```

```python
y_pred = classifier.predict(X_test)
y_pred = (y_pred >.5)

#confusion matrix
cm=confusion_matrix(y_test,y_pred)
#Performance metrics:
#Accuracy
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy is %f' % accuracy)
# Precision
precision = precision_score(y_test, y_pred)
print('Precision is %f' % precision)
# Recall
recall = recall_score(y_test, y_pred)
print('Recall is %f' % recall) 
```
